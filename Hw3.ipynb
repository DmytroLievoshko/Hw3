Построение узора Серпинского:

```python
import random
import matplotlib.pyplot as plt

vertices = [(0, 0), (1, 0), (0.5, 0.87)]

point = (0.5, 0.5)

num_iterations = 100000

points = [point]

for i in range(num_iterations):
    vertex = random.choice(vertices)
    point = ((point[0]+vertex[0])/2, (point[1]+vertex[1])/2)
    points.append(point)

plt.figure(figsize=(10, 10))
plt.scatter([p[0] for p in points], [p[1] for p in points], s=0.1)
plt.axis('equal')
plt.show()
```

Зависимости доходности акций

```python
import seaborn as sns
import pandas as pd

data = {'Доходность A %': [25, -10, 10, 5, 35, 13],
        'Доходность B %': [0, 15, -5, 5, 20, 25],
        'Доходность C %': [10, 25, -15, -5, -5, 15]}

df = pd.DataFrame(data)
```

доходность акции A от доходности акции B

```python
sns.regplot(x='Доходность B %', y='Доходность A %', data=df)
plt.title('Зависимость доходности акции A от доходности акции B')
plt.xlabel('Доходность B %')
plt.ylabel('Доходность A %')
plt.show()
```

доходность акции A от доходности акции C

```python
sns.regplot(x='Доходность C %', y='Доходность A %', data=df)
plt.title('Зависимость доходности акции A от доходности акции C')
plt.xlabel('Доходность C %')
plt.ylabel('Доходность A %')
plt.show()
```

доходность акции C от доходности акции B

```python
sns.regplot(x='Доходность B %', y='Доходность C %', data=df)
plt.title('Зависимость доходности акции C от доходности акции B')
plt.xlabel('Доходность B %')
plt.ylabel('Доходность C %')
plt.show()
```

коэффициенты корреляции

```python
corr_matrix = df.corr().abs()
min_corr = corr_matrix[corr_matrix != 1].min().min()
pair = corr_matrix[(corr_matrix != 1) & (corr_matrix == min_corr)].stack().index.tolist()
least_dependent = pair[0], pair[1]
print("Наименее зависимая пара акций:", least_dependent)
```
